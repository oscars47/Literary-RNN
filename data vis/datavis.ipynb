```python
# imports
import matplotlib.pyplot as plt
import pandas as pd
```

## Read in Data, clean csv

```python
# read in csv
data_raw = pd.read_csv('survey_12_25.csv')
data_raw
```

```python
# print out columns
data_raw.columns
```

```python
# initialize clean df
data_clean = pd.DataFrame()
# list of possible ratings
ratings_key = ['Poor', 'Fair', 'Adequate', 'Good', 'Exceptional']
creative1=[]
creative2=[]
creative3=[]
creative4=[]
cohesive1=[]
cohesive2=[]
cohesive3=[]
cohesive4=[]
compelling1=[]
compelling2=[]
compelling3=[]
compelling4=[]
for i in range(len(data_raw)): # for every reviewer
    # go through every object
    
    creative1.append(ratings_key.index(data_raw['1creative'][i].split(':')[0]))
    cohesive1.append(ratings_key.index(data_raw['1cohesive'][i].split(':')[0]))
    compelling1.append(ratings_key.index(data_raw['1compelling'][i].split(':')[0]))

    creative2.append(ratings_key.index(data_raw['2creative'][i].split(':')[0]))
    cohesive2.append(ratings_key.index(data_raw['2cohesive'][i].split(':')[0]))
    compelling2.append(ratings_key.index(data_raw['2compelling'][i].split(':')[0]))

    creative3.append(ratings_key.index(data_raw['3creative'][i].split(':')[0]))
    cohesive3.append(ratings_key.index(data_raw['3cohesive'][i].split(':')[0]))
    compelling3.append(ratings_key.index(data_raw['3compelling'][i].split(':')[0]))

    creative4.append(ratings_key.index(data_raw['4creative'][i].split(':')[0]))
    cohesive4.append(ratings_key.index(data_raw['4cohesive'][i].split(':')[0]))
    compelling4.append(ratings_key.index(data_raw['4compelling'][i].split(':')[0]))

data_clean['Age'] = data_raw['Age']
data_clean['Creative1'] = creative1
data_clean['Creative2'] = creative2
data_clean['Creative3'] = creative3
data_clean['Creative4'] = creative4
data_clean['Cohesive1'] = cohesive1
data_clean['Cohesive2'] = cohesive2
data_clean['Cohesive3'] = cohesive3
data_clean['Cohesive4'] = cohesive4
data_clean['Compelling1'] = compelling1
data_clean['Compelling2'] = compelling2
data_clean['Compelling3'] = compelling3
data_clean['Compelling4'] = compelling4

data_clean.to_csv('data_clean.csv')
data_clean
```

## Making plots

```python
# visualizing histogram of ages
ages= data_clean['Age']
plt.figure(figsize=(10,7))
counts, edges, bars = plt.hist(ages, color='green', alpha=0.8, bins=50)
plt.bar_label(bars)# add count labels
plt.xlabel('Age', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.title('Age distribution for N='+str(len(data_clean))+' respondents', fontsize=18)
plt.show()
```

```python
def autopct_format(values):
        def my_format(pct):
            total = sum(values)
            val = int(round(pct*total/100.0))
            return '{:.1f}%'.format(pct, v=val)
        return my_format

# for pie charts; name is type of thing being rated
def make_4_pie(list1, list2, list3, list4, name):
    figure, axis = plt.subplots(2, 2)

    colors = {'Poor': 'violet',
           'Fair': 'magenta',
           'Adequate': 'Orange',
           'Good': 'Red',
           'Exceptional': 'Cyan'}

    # get unique
    list1_unique = set(list1)
    list1_labels = []
    list1_count = []
    for uni in list1_unique:
        list1_count.append(list1.count(uni))
        list1_labels.append(ratings_key[uni])

    axis[0, 0].pie(list1_count, labels=list1_labels, autopct=autopct_format(list1), colors=[colors[key] for key in list1_labels])
    axis[0,0].set_title('Text 1')

    # get unique
    list2_unique = set(list2)
    list2_labels = []
    list2_count = []
    for uni in list2_unique:
        list2_count.append(list2.count(uni))
        list2_labels.append(ratings_key[uni])

    print(len(list2_labels))

    axis[0, 1].pie(list2_count, labels=list2_labels, autopct=autopct_format(list2), colors=[colors[key] for key in list2_labels])
    axis[0,1].set_title('Text 2')

    # get unique
    list3_unique = set(list3)
    list3_labels = []
    for label in list3_unique:
        list3_labels.append(ratings_key[label])
    list3_count = []
    for uni in list3_unique:
        list3_count.append(list3.count(uni))

    axis[1, 0].pie(list3_count, labels=list3_labels, autopct=autopct_format(list3), colors=[colors[key] for key in list3_labels])
    axis[1,0].set_title('Text 3')

    # get unique
    list4_unique = set(list4)
    list4_labels = []
    for label in list4_unique:
        list4_labels.append(ratings_key[label])
    list4_count = []
    for uni in list4_unique:
        list4_count.append(list4.count(uni))

    axis[1,1].pie(list4_count, labels=list4_labels, autopct=autopct_format(list4), colors=[colors[key] for key in list4_labels])
    axis[1,1].set_title('Text 4')

    figure.suptitle(name+', N='+str(len(list1)), fontsize=18)
    plt.savefig(name+'.jpeg')
    plt.show()
```

```python
# function to make plots for all texts
def make_all_plots(data_clean, name_list):
    make_4_pie(data_clean['Creative1'].to_list(), data_clean['Creative2'].to_list(), data_clean['Creative3'].to_list(), data_clean['Creative4'].to_list(), name_list[0])
    make_4_pie(data_clean['Cohesive1'].to_list(), data_clean['Cohesive2'].to_list(), data_clean['Cohesive3'].to_list(), data_clean['Cohesive4'].to_list(), name_list[1])
    make_4_pie(data_clean['Compelling1'].to_list(), data_clean['Compelling2'].to_list(), data_clean['Compelling3'].to_list(), data_clean['Compelling4'].to_list(), name_list[2])
```

```python
# get plots for all
make_all_plots(data_clean, ['Creativity', 'Cohesive', 'Compelling'])
```

```python
# now get restricted
data_clean_restricted = data_clean.loc[data_clean['Age'] < 30]
make_all_plots(data_clean_restricted, ['CreaClip', 'CohClip', 'ComClip'])
```

```python
# combine all plots
figure, axis = plt.subplots(6, 4)

colors = {'Poor': 'violet',
        'Fair': 'magenta',
        'Adequate': 'Orange',
        'Good': 'Red',
        'Exceptional': 'Cyan'}


data_ls=[[data_clean['Creative1'].to_list(), data_clean['Creative2'].to_list(), data_clean_restricted['Creative1'].to_list(), data_clean_restricted['Creative2'].to_list()],
[data_clean['Creative3'].to_list(), data_clean['Creative4'].to_list(), data_clean_restricted['Creative3'].to_list(), data_clean_restricted['Creative4'].to_list()],
[data_clean['Cohesive1'].to_list(), data_clean['Cohesive2'].to_list(), data_clean_restricted['Cohesive1'].to_list(), data_clean_restricted['Cohesive2'].to_list()],
[data_clean['Cohesive3'].to_list(), data_clean['Cohesive4'].to_list(), data_clean_restricted['Cohesive3'].to_list(), data_clean_restricted['Cohesive4'].to_list()],
[data_clean['Compelling1'].to_list(), data_clean['Compelling2'].to_list(), data_clean_restricted['Compelling1'].to_list(), data_clean_restricted['Compelling2'].to_list()],
[data_clean['Compelling3'].to_list(), data_clean['Compelling4'].to_list(), data_clean_restricted['Compelling3'].to_list(), data_clean_restricted['Compelling4'].to_list()]]

name_ls = [['Crea1', 'Crea2', 'CreaClip1', 'CreaClip2'],
['Crea3', 'Crea4', 'CreaClip3', 'CreaClip4'],
['Coh1', 'Coh2', 'CohClip1', 'CohClip2'],
['Coh3', 'Coh4', 'CohClip3', 'CohClip4'],
['Com1', 'Com2', 'ComClip1', 'ComClip2'],
['Com3', 'Com4', 'ComClip3', 'ComClip4']]

for j in range(len(data_ls)):
    ls = data_ls[j]
    list1 = ls[0]
    list2 = ls[1]
    list3 = ls[2]
    list4 = ls[3]

    name = name_ls[j]
    text1 = name[0]
    text2 =name[1]
    text3=name[2]
    text4=name[3]

    # get unique
    list1_unique = set(list1)
    list1_labels = []
    list1_count = []
    for uni in list1_unique:
        list1_count.append(list1.count(uni))
        list1_labels.append(ratings_key[uni])

    axis[j, 0].pie(list1_count, labels=list1_labels, autopct=autopct_format(list1), colors=[colors[key] for key in list1_labels])
    axis[j,0].set_title(text1)

    # get unique
    list2_unique = set(list2)
    list2_labels = []
    list2_count = []
    for uni in list2_unique:
        list2_count.append(list2.count(uni))
        list2_labels.append(ratings_key[uni])

    print(len(list2_labels))

    axis[j, 1].pie(list2_count, labels=list2_labels, autopct=autopct_format(list2), colors=[colors[key] for key in list2_labels])
    axis[j,1].set_title(text2)

    # get unique
    list3_unique = set(list3)
    list3_labels = []
    for label in list3_unique:
        list3_labels.append(ratings_key[label])
    list3_count = []
    for uni in list3_unique:
        list3_count.append(list3.count(uni))

    axis[j, 2].pie(list3_count, labels=list3_labels, autopct=autopct_format(list3), colors=[colors[key] for key in list3_labels])
    axis[j,2].set_title(text3)

    # get unique
    list4_unique = set(list4)
    list4_labels = []
    for label in list4_unique:
        list4_labels.append(ratings_key[label])
    list4_count = []
    for uni in list4_unique:
        list4_count.append(list4.count(uni))

    axis[j,3].pie(list4_count, labels=list4_labels, autopct=autopct_format(list4), colors=[colors[key] for key in list4_labels])
    axis[j,3].set_title(text4)

figure.suptitle('Results: N='+str(len(data_ls[0][0])) + ", N'="+str(len(data_ls[0][2])), fontsize=18)
plt.savefig('complete_results.jpeg')
plt.show()
```

```python
# getting total scores for each metric
crea_scores=[]
cohe_scores=[]
comp_scores=[]


crea_ls = [data_clean['Creative1'].to_list(), data_clean['Creative2'].to_list(),
data_clean['Creative3'].to_list(), data_clean['Creative4'].to_list()]
cohe_ls = [data_clean['Cohesive1'].to_list(), data_clean['Cohesive2'].to_list(),
data_clean['Cohesive3'].to_list(), data_clean['Cohesive4'].to_list()]
comp_ls = [data_clean['Compelling1'].to_list(), data_clean['Compelling2'].to_list(),
data_clean['Compelling3'].to_list(), data_clean['Compelling4'].to_list()]


total_ls = [crea_ls, cohe_ls, comp_ls]
for i, metric_ls in enumerate(total_ls):
    for ls in metric_ls:
        score = sum(ls)/(4*len(ls))
        if i == 0:
            crea_scores.append(score)
        elif i==1:
            cohe_scores.append(score)
        elif i == 2:
            comp_scores.append(score)

plt.figure(figsize=(10,7))
plt.plot(list(range(4)), crea_scores, label='Creativity')
plt.plot(list(range(4)), cohe_scores, label='Cohesion')
plt.plot(list(range(4)), comp_scores, label='Compellingness')

plt.legend()
plt.xlabel('Excerpt', fontsize=16)
plt.ylabel('Complete Scores', fontsize=16)
plt.title('Scores Per Excerpt', fontsize=18)
plt.show()
```
